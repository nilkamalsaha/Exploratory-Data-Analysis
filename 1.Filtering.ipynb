```python
import pandas as pd
groceries = pd.read_pickle('groceries_with_new_columns.pkl')
groceries.head()
```

```python
# filtering Low Inventory column with only Low Inventory tag
groceries.loc[groceries['Low Inventory'] == 'Low Inventory']
```

```python
# filtering Low Inventory column with only Low Inventory tag (only one column so without loc possible but if 2 columns , there must be  .loc to use)
groceries[groceries['Low Inventory'] == 'Low Inventory']
```

```python
# filtering Low Inventory column with only Low Inventory tag and price of those and Inventory column of those as well
groceries.loc[groceries['Low Inventory'] == 'Low Inventory', ['Price_Dollars', 'Inventory']]

```

```python
# before filtering, remove if there is any space and save it to Data Frame. after saving, it will be reflected following steps
groceries.Subcategory = groceries.Subcategory.str.strip()
groceries.Subcategory[17] # there is no space before Snacks
```

```python
# filtering Subcategory column with only Snacks tag (there was space before snacks)
groceries.loc[groceries['Subcategory'] == 'Snacks']
```

```python
# filtering 2 columns Low Inventory with Low Inventory tag only and Subcategory with Snacks tag only
mask = (groceries['Low Inventory'] == 'Low Inventory') & (groceries['Subcategory'] == 'Snacks')
```

```python
groceries[mask]
```

```python
# filtering only 'Subcategory', 'Item', 'Inventory' columns with 'Low Inventory' and 'Snacks' tags only
groceries.loc[mask, ['Subcategory', 'Item', 'Inventory']]
```

```python
import pandas as pd
groceries = pd.read_pickle('groceries_with_new_columns.pkl')
groceries.head()
```

```python
#To group the Inventory column based on Category column 
groceries.groupby('Category')['Inventory'].sum()
```

```python
#To filter how many category are available
groceries['Category'].value_counts()
```

```python
#To group the Inventory column based on Category column 
groceries.groupby('Category')['Inventory'].count()
```

```python
#To group the Inventory column based on Category and Subcategory column 
groceries.groupby(['Category', 'Subcategory'])['Inventory'].count()
```

```python
#To group the Inventory column based on Category and Subcategory column as well as to set to a DataFrame
groceries.groupby(['Category', 'Subcategory'])['Inventory'].count().reset_index()
```

```python
#To group the Inventory column based on Category and Subcategory column as well as to set to a DataFrame
groceries.groupby(['Category', 'Subcategory'])['Inventory'].agg(['sum', 'count']).reset_index()
```

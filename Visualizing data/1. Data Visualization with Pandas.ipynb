```python
import pandas as pd
groceries = pd.read_pickle('groceries_with_new_columns.pkl')
groceries.head()
```

```python
#creating group of Subcategory price
groceries.groupby(['Subcategory'])['Price_Dollars'].mean()
```

```python
#creating bar plot chart by grouping of Subcategory price
groceries.groupby(['Subcategory'])['Price_Dollars'].mean().plot.bar();
```

```python
#creating horizontal bar plot chart by grouping of Subcategory price
groceries.groupby(['Subcategory'])['Price_Dollars'].mean().plot.barh();
```

```python
#creating horizontal bar plot chart by grouping of Subcategory price in ascending order
groceries.groupby(['Subcategory'])['Price_Dollars'].mean().sort_values().plot.barh();
```

```python
happiness_survey = pd.read_csv('happiness_survey_data.csv')
happiness_survey.head()
```

```python
#creating  a line chart of happiness score of mexico over time (year and happiness score column only) 

happiness_survey[happiness_survey.country_name == 'Mexico']
```

```python
# year and happiness score column only
happiness_survey[happiness_survey.country_name == 'Mexico'].iloc[:, 1:3]
```

```python
#creating a line plot
happiness_survey[happiness_survey.country_name == 'Mexico'].iloc[:, 1:3].plot.line(x= 'year', y='happiness_score')
```

```python
#creating multiple line charts for different countries
(happiness_survey[happiness_survey.country_name
                .isin(['Mexico', 'Canada', 'United States'])]
                .iloc[:, :3]
                .pivot(index= 'year', columns = 'country_name', values='happiness_score')
                .plot.line())
```
